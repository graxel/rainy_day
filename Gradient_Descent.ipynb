```python
rcParams['figure.figsize'] = (12, 3)
x = np.array([0,1,2])
y = np.array([2,1,0])
plt.scatter(x, y, c = 'r')
plt.show()
```

```python
def gradient_descent(x,y, alpha, theta0 = 0, theta1 = 0, iterations = 20):
    rcParams['figure.figsize'] = (16, 10)
    costs = []
    theta0s = [theta0]
    theta1s = [theta1]
    plt.scatter(x, y, c = 'r')
    y_pred = theta0 + theta1*x
    mse = np.mean(np.power((y_pred - y), 2))
    costs.append(mse)
    for i in range(iterations):
        plt.plot(x, y_pred, '--', label = i)
        temp_theta0 = theta0 - (alpha / len(points)) * sum(theta0 + theta1*x - y)
        theta1 = theta1 - (alpha / len(points)) * sum((theta0 + theta1*x - y)*x)
        theta0 = temp_theta0 
        theta0s.append(theta0)
        theta1s.append(theta1)
        y_pred = theta0 + theta1*x
        mse = np.mean(np.power((y_pred - y), 2))
        costs.append(mse)

    plt.legend(loc = 'upperright', title = 'Iteration')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gradient Descent for each Iteration')
    plt.show()
    
    plt.plot(range(i + 2), costs)
    plt.xticks(range(i+2))
    plt.xlabel('Iteration')
    plt.ylabel('MSE')
    plt.title('Mean Squared Error for each Iteration')
    
    
    return theta0s, theta1s, costs 
```

```python
rcParams['axes.facecolor'] = '#eeeeee'
theta0s, theta1s, costs = gradient_descent(x,y,alpha = 0.5)
```

```python
X,Y = np.meshgrid(np.array(theta0s), np.array(theta1s))
Z = np.array([np.array(costs) for i in range(len(costs))])
```

```python
rcParams['axes.facecolor'] = 'white'
rcParams['grid.color'] = 'white'
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X,Y, Z)

plt3d.set_xlabel('$\Theta_{0}$', fontsize=20)
plt3d.set_ylabel('$\Theta_{1}$', fontsize=20)
plt3d.set_zlabel('$J(\Theta)$', fontsize=20)

for t in plt3d.xaxis.get_major_ticks(): t.label.set_fontsize(10)
for t in plt3d.yaxis.get_major_ticks(): t.label.set_fontsize(10)
for t in plt3d.zaxis.get_major_ticks(): t.label.set_fontsize(10)


plt.show()
```
